```python
df_user['creation_date'] = pd.to_datetime(df_user['creation_date'],errors='coerce',format='%Y-%m-%dT%H:%M:%S.%f')
df_post['creation_date'] = pd.to_datetime(df_post['creation_date'],errors='coerce',format='%Y-%m-%dT%H:%M:%S.%f')
df_post['closed_date'] = pd.to_datetime(df_post['closed_date'],format='%Y-%m-%dT%H:%M:%S.%f')
df_post['community_owned_date'] = pd.to_datetime(df_post['community_owned_date'],format='%Y-%m-%dT%H:%M:%S.%f')

df_ph['creation_date'] = pd.to_datetime(df_ph['creation_date'],format='%Y-%m-%dT%H:%M:%S.%f')
```

```python
df_answers =df_post[df_post['post_type_id']==2]
df_questions = df_post[df_post['post_type_id']==1]
```

```python
id=[]
dateTime =[]
for rows,values in df_answers.groupby('parent_id').groups.items():
  id.append(rows)
  dateTime.append(df_answers.loc[values[0]]["creation_date"])
```

```python
import pandas as pd
filtered_ans = pd.DataFrame({'id':id, 'Answer_Time':dateTime})
```

```python
final_df = pd.merge(df_questions,filtered_ans,on=['id'])
```

```python
import re
def convertTagsToList(input_tags):
    output = []
    for tag in re.split("<|><|>",input_tags):
        if tag!='':
            output.append(tag)
    return output
tag_dict = {}
count = 0
id_List = []
for (i,row) in tags_data.iterrows():
    #tags_data['tags'].apply(str)
    id = row['id']
    id_List.append(id)
    tag_list = convertTagsToList(row['tags'])
    count = count +1
    for tag in tag_list:
        if tag in tag_dict:
            tag_dict[tag] = tag_dict[tag]+1
        else:
            tag_dict[tag] = 1
print(tag_dict) 
print(len(tag_list))  
```
